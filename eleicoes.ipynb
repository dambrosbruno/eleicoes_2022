```python
import pandas as pd
import plotly.express as px
```

```python
df_sc = pd.read_csv(
    filepath_or_buffer = r'C:\Users\dambr\Documents\eleicoes\logs\bweb_2t_SC_311020221535.csv',
    sep=';',
    encoding='cp1252',
    usecols=['DS_CARGO_PERGUNTA', 'NM_MUNICIPIO', 'NR_ZONA', 'NR_SECAO', 'NR_LOCAL_VOTACAO', 'NR_URNA_EFETIVADA', 'QT_APTOS', 'QT_COMPARECIMENTO', 'QT_ABSTENCOES', 'NM_VOTAVEL', 'QT_VOTOS', 'DT_EMISSAO_BU']) 
```

```python
# Deletar os governadores
mask = df_sc['DS_CARGO_PERGUNTA'].isin(['Governador'])
df_sc = df_sc[~mask]
```

```python
# Deletar linhas duplicadasf
df_sc = df_sc.drop_duplicates()
```

```python
#Converter coluna de data de emissão do BU para DateTime
df_sc['DT_EMISSAO_BU'] =  pd.to_datetime(df_sc['DT_EMISSAO_BU'])
```

```python
#Pegar só o time do DT_EMISSAO_BU
df_sc['HR_EMISSAO_BU'] = df_sc['DT_EMISSAO_BU'].dt.strftime('%H:%M:%S')
```

```python
#Total de votos em SC no 2º turno segundo:
total = df_sc['QT_VOTOS'].sum() #4842282
total
```

```python
# Total de votos por candidato
df_votos = df_sc.groupby('NM_VOTAVEL')['QT_VOTOS'].sum()
df_votos
```

```python
df_sc.groupby(['NM_VOTAVEL', 'NM_MUNICIPIO'], as_index=True, sort=True)['QT_VOTOS'].sum()
```

```python
#Coverter HR_EMISSAO_BU em time
df_sc['HR_EMISSAO_BU'] = pd.to_datetime(df_sc['HR_EMISSAO_BU'], unit='s')
```
